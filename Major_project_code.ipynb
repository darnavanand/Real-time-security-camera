```python
import os
import pandas as pd
import random
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import cv2 as cv2
```

```python
#DIRECTORY is the location of folder where our family member folder and Unknown_faces folder are present
DIRECTORY = r"C:\\Users\\Arnav Anand\\OneDrive\\Documents\\faces"
#CATEGORIES are array of two items and it is the name of the sub-folder
CATEGORIES = ['family member','Unknown_faces']

count = 0

#data is array in which all the images will be stored
data = []

#target is the array in which 0 for family member and 1 for Unknown_faces will be stored
target =[]

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY,category)
    label = CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        gray = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)
        resize_image = cv2.resize(gray, (150, 150))
        #here data.append means each image is stored one by one in data array
        data.append(resize_image)
        #here target.append means for each image 0 or 1 is stored in target array
        target.append(count)
        
    count+=1
    
```

```python
#gives number of images present in data array
len(data)
```

```python
#normalizing and reshaping all the images present in data array
data_normalized = np.array(data)/255.0
data_normalized_reshaped = data_normalized.reshape(207,150,150,1)
```

```python
data_normalized_reshaped
```

```python
import tensorflow
from keras.utils import np_utils
target = np_utils.to_categorical(target)
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_normalized_reshaped, target)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Dropout, Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
model.add(Conv2D(300, (3,3), input_shape=(150,  150, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(100, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.3))
model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
cp = ModelCheckpoint('best_one', verbose=0, save_best_only=True)
model.fit(X_train, Y_train, epochs = 5, callbacks=[cp], validation_split=0.2)
```

```python
#importing these libraries for email

import os
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email.mime.image import MIMEImage
from email import encoders
```

```python
source = cv2.VideoCapture('obama.mp4')

while 1:        
    _, image = source.read(0)
    #print("Captured image")
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    detected_faces = face_detect.detectMultiScale(gray, 1.2, 9)
    #print("Detected faces",len(detected_faces))
    if len(detected_faces) > 1:
        continue
    for each_face in detected_faces:
        # Each bbox is a rectangle representing
        # the bounding box around the detected object
        x, y, w, h = each_face
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 0, 255), 3)
        # Cropped object
        crop = gray[y:y+h, x:x+w]
        cropped_image = cv2.resize(crop, (150, 150))
        normalized_image = cropped_image/255
        reshaped_face = np.reshape(normalized_image, (1, 150, 150, 1))
        result = model.predict(reshaped_face)[0]
        if result[0] >  result[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(image, "family member", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
        
        if result[1] >  result[0]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
            cv2.putText(image, "intruder", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            
            #these line of codes is for saving image captured by webcam
            sub_face = image[y:y+h, x:x+w]
            
            #this is the location where image is saved in .jpg extension
            FaceFileName = "C:\\Users\Arnav Anand\\OneDrive\\Documents\\Captured faces" + str(y) + ".jpg"
            
            #time.sleep is used so that detected image is captured and saved after every 5 sec
            time.sleep(5)
            cv2.imwrite(FaceFileName, sub_face)
            
            #these codes are for emailing 
            msg = MIMEMultipart()
            img_data = open(FaceFileName, 'rb').read()
            msg['Subject'] = 'Intruder'
            msg['From'] = 'from_email_address'
            msg['To'] = 'To_email_address'

            message = "Someone Entered Your Room"
            msg.attach(MIMEText(message, 'plain'))
            img = MIMEImage(img_data, name=os.path.basename(FaceFileName))               
            msg.attach(img)
            s = smtplib.SMTP_SSL('smtp.googlemail.com' , 465)
            s.ehlo()
            
            
            s.login(msg['From'] , 'Password')
            s.sendmail(msg['From'], msg['To'], msg.as_string())
            
            

                            

                            
            
    cv2.imshow('Verzeo Project', image)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
source.release()
```
